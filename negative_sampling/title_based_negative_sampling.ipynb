Title-based Hard Negative Sampling

```python
import pandas as pd
from datasets import load_dataset
KorQuad = load_dataset('KETI-AIR/korquad','v1.0')
```

1. KorQuad로 6000개 hard negative 모의 실험
2. klue로 hard negative 실험

```python
df = pd.DataFrame(KorQuad['train'])
```

```python
df.head(20)
```

KorQuad data에 대해 negative sampling 진행

```python
# KorQuad data에 대해 negative sampling
import random
from tqdm.notebook import tqdm

negative_sampled = pd.DataFrame(KorQuad['train'])

unique_title = df['title'].unique()
for title in tqdm(unique_title, desc="negative sampling"):
    possi_df = df[df['title'] == title] #title이 같은 곳에서 변경
    unique_context = possi_df['context'].unique()
    if len(unique_context) >= 3: # unique 개수가 3 이상일 때 진행
        for pos, start, id in zip(possi_df['context'], possi_df['answers'], possi_df['id']):
            false_list = []
            for context in unique_context: #각각의 context에 대해 진행
                if context != pos:
                    false_list.append(context)
            #2개 뽑아오기
            chosen_false = random.sample(false_list, 2)
            r =  random.randint(1,3) #랜덤하게 수 지정하기
            if r == 1: #FTF
                for i, v in enumerate(chosen_false):
                    if i == 0:
                        pos = v + pos
                        s = start['answer_start'][0] + len(v)
                    else:
                        pos = pos + v
            elif r == 2: #FFT
                s = start['answer_start'][0]
                for i, v in enumerate(chosen_false):
                    pos = v + pos
                    s = s + len(v)
            else: #TFF
                s = start['answer_start'][0]
                for i, v in enumerate(chosen_false):
                    pos = pos + v
            negative_sampled.loc[negative_sampled['id']==id, 'context'] = pos
            negative_sampled[negative_sampled['id']==id].iloc[0, 4]['answer_start'][0] = s
```

```python
negative_sampled.to_csv('negative_sampled_KoQuad.csv', index=False)
```

KLUE data에 대해 negative sampling 진행

```python
from datasets import load_from_disk
klue = load_from_disk('./data/train_dataset')
```

```python
# klue data에 대해 negative sampling
import random
import pandas as pd
from tqdm.notebook import tqdm

df = pd.DataFrame(klue['train'])
negative_sampled = pd.DataFrame(klue['train'])

unique_title = df['title'].unique()
for title in tqdm(unique_title, desc="negative sampling"):
    possi_df = df[df['title'] == title] #title이 같은 곳에서 변경
    unique_context = possi_df['context'].unique()
    if len(unique_context) >= 2: # unique 개수가 2 이상일 때 진행
        for pos, start, id in zip(possi_df['context'], possi_df['answers'], possi_df['id']):
            false_list = []
            for context in unique_context: #각각의 context에 대해 진행
                if context != pos:
                    false_list.append(context)
            #2개 뽑아오기
            chosen_false = random.sample(false_list, min(2, len(false_list)))
            for i, v in enumerate(chosen_false):
                r = random.randint(1,2) #랜덤하게 수 지정하기
                s = 0
                # 11 -> FFT, 12 -> FTF, 21 -> FTF, 22 -> TFF
                if r == 1:
                    pos = v + pos
                    s = s + len(v)
                else: #TF
                    pos = pos + v
            negative_sampled.loc[negative_sampled['id']==id, 'context'] = pos
            negative_sampled[negative_sampled['id']==id].iloc[0, 4]['answer_start'] = [x+s for x in start['answer_start']]
```

```python
negative_sampled.to_csv("negative_sampled_klue.to_csv", index=False)
```

허깅페이스 데이터 형태로 바뀐 KLUE data 저장

```python
# klue data df를 허깅페이스 데이터 형태로 저장
import pandas as pd
import datasets
from datasets import Dataset, DatasetDict
tds = Dataset.from_pandas(negative_sampled)
del klue['train']
klue['train'] = tds
klue.save_to_disk("klue_negative.hf")
```
